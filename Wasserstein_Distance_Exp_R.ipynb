```r
library(repr)
library(reticulate)

np <- import("numpy")
```

```r
wass_mat = np$load("Exp_Wasserstein_Pairwise.npy")
```

```r
npz = np$load("wasserstein_matrix_exp.npz")

wass_mat_alt = npz$f[["wassmatrix"]]
densities = npz$f[["densities"]]
well_nums = npz$f[["well_nums"]]


conditions = c("GM", "Gefitinib", "Assay", "GM", "Gefitinib", "Assay",
             "Assay", "Gefitinib", "GM", "Assay", "Gefitinib", "GM",
             "GM", "Gefitinib", "Assay", "GM", "Gefitinib", "Assay",
             "Assay", "Gefitinib", "GM", "Assay", "Gefitinib", "GM")


hc = hclust(as.dist(wass_mat), "complete")
```

```r
leaves_labels = list()

for (l in 1:length(hc$order)){
    well_label = ""
    if (well_nums[l] < 25){
        well_label = "DMSO, "
    } else {
        well_label = "OHT, "
    }
    
    cond_str = paste(well_label, conditions[l], ", ", densities[l], " cells (Well ", well_nums[l], ")", sep="")
    leaves_labels = c(leaves_labels, cond_str)
}
```

```r
options(repr.plot.width=30, repr.plot.height=10)
plot(hc, label=leaves_labels)
```
