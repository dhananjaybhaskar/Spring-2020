```r
library(ggplot2)
library(reticulate)
np <- import("numpy")
```

```r
clustered = np$load("clustered_removal.npy")
clustered_prof = np$load("clustered_proliferation_removal.npy")
spanning = np$load("spanning_removal.npy")
spanning_prof = np$load("spanning_proliferation_removal.npy")
individual = np$load("individual_removal.npy")
individual_prof = np$load("individual_proliferation_removal.npy")
```

```r
plaw = function(x,a,b){a*x^(b)}
```

#### Spanning (no proliferation)

```r
xdata_spanning = c(seq(1, 180), seq(1, 180), seq(1, 180), seq(1, 180), seq(1, 180))
ydata_spanning = c(spanning[,1], spanning[,2], spanning[,3], spanning[,4], spanning[,5])
```

```r
options(repr.plot.width = 8, repr.plot.height = 4)
par(mfrow=c(1,2))

plot(xdata_spanning, ydata_spanning)
curve(plaw(x, 0.1640, 0.5277), add=TRUE, col="red")

plot(log10(xdata_spanning), log10(ydata_spanning), type="p", 
     xlab="Log(# Removals)", ylab="Log(Wasserstein Distance)", 
    cex=0.5, pch=21, col=rgb(0,0,0,alpha=0.2), bg=rgb(0,0,0,alpha=0.2))
```

```r
fit = nls(ydata_spanning~plaw(xdata_spanning, a, b),start=list(a=0.2,b=0.57))
print(fit)
```

```r
# Expect 1.75 (OHT)
1/0.5277
```

```r
spanning_mean = apply(spanning, 1, mean)
spanning_std = apply(spanning, 2, sd)
clustered_mean = apply(clustered, 1, mean)
clustered_std = apply(clustered, 2, sd)
individual_mean = apply(individual, 1, mean)
individual_std = apply(individual, 2, sd)

spanning.func <- function(x) 0.1640*x^0.5277
clustered.func <- function(x) 0.2403*x^0.3450 
individual.func <- function(x) 0.2305*x^0.5232
```

```r
options(repr.plot.width = 5, repr.plot.height = 4)
ggplot() + 
geom_point(aes(x=seq(1, 180), y=spanning_mean), color='red', alpha=0.3) +
geom_errorbar(aes(x=seq(3, 180), ymin=(spanning_mean-0.5*spanning_std)[3:180], ymax=(spanning_mean+0.5*spanning_std)[3:180]), color='red', alpha=0.3, width=0.02) +
geom_point(aes(x=seq(1, 180), y=clustered_mean), color='blue', alpha=0.3) +
geom_errorbar(aes(x=seq(3, 180), ymin=(clustered_mean-0.5*clustered_std)[3:180], ymax=(clustered_mean+0.5*clustered_std)[3:180]), color='blue', alpha=0.3, width=0.02) +
geom_point(aes(x=seq(1, 180), y=individual_mean), color='darkgreen', alpha=0.3) +
geom_errorbar(aes(x=seq(3, 180), ymin=(individual_mean-0.5*individual_std)[3:180], ymax=(individual_mean+0.5*individual_std)[3:180]), color='darkgreen', alpha=0.3, width=0.02) +
stat_function(data = data.frame(x = 0), mapping = aes(x = x), fun = spanning.func, color='red') +
stat_function(data = data.frame(x = 0), mapping = aes(x = x), fun = clustered.func, color='blue') +
stat_function(data = data.frame(x = 0), mapping = aes(x = x), fun = individual.func, color='darkgreen') +
theme_bw() + scale_x_continuous(trans='log10') + scale_y_continuous(trans='log10') +
xlab('Log(Number of Removals)') + ylab('Log(Wasserstein Distance)')
```
