```julia
filepath = eirenefilepath("simplecity")
a = ezread(filepath)
b = a[2:end,3:4]
```

```julia
b = convert(Array{Float64,2},b)
```

```julia
c = latlon2euc(b,model="points") 
```

```julia
d = ezlabel(a[2:end,2])   
```

```julia
C = eirene(c,model= "pc",maxdim = 1, maxrad = 0.15, pointlabels=d)
```
